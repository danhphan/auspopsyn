```python
import numpy as np
import pandas as pd
```

```python
# Read data
file_path = "../data/melbourne/vista/raw/H_VISTA12_16_SA1_V1.csv"
hholds = pd.read_csv(file_path, low_memory=False)
hholds.shape
```

### Process Households

```python
# Remove NAN values
hholds = hholds.dropna()
hholds.shape
```

```python
hholds.head(10)
```

```python
hholds.describe()
```

```python
hholds.HOMEPC = hholds.HOMEPC.astype('object')
```

```python
hholds.select_dtypes(include='object').describe()
```

```python
# Rename
map_cols = {'HHID':'HHID','TRAVDOW':'TravelDay','TRAVMONTH':'TravelMonth','TRAVYEAR':'TravelYear','OWNDWELL':'OwnDwell',
            'DayType':'DayType','HHINC':'HhIncome','HHSIZE':'HhSize','MONTHSLIVED':'MonthsLived','YEARSLIVED':'YearsLived',
            'VISITORS':'Visitors','KIDSBIKES':'KidBikes','ADULTBIKES':'AdultBikes','TOTALBIKES':'TotalBikes',
            'MBIKES':'TotalMotoBikes','CARS':'TotalCars','TOTALVEHS':'TotalVehicles','FOURWDS':'Num4WDs','TRUCKS':'Trucks',
            'UTES':'Utes','VANS':'Vans','OTHERVEHS':'OtherVehs','HOMELGA':'HomeLGA','HOMEPC':'HomePostcode'}
hholds = hholds.rename(columns=map_cols)
```

```python
# Processing Household Income
hholds.HhIncome = hholds.HhIncome.str.replace("$","").str.replace(",","").str.replace(' ','')
```

```python
hholds.HhIncome = pd.to_numeric(hholds.HhIncome,errors='raise')
```

```python
hholds.HhIncome.fillna(0,inplace=True)
```

```python
# Processing travel day and month
hholds['TravelMonthNum'] = hholds.TravelMonth.map({'January':1,'February':2,'March':'3','April':'4','May':5,'June':6,
                                               'July':7,'August':8,'September':9,'October':10,'November':11,'December':12})
hholds['TravelDayNum'] = hholds.TravelDay.map({'Monday':2,'Tuesday':3,'Wednesday':'4','Thursday':'5',
                                               'Friday':6,'Saturday':7,'Sunday':8})
```
